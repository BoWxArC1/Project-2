```python
# What percent of colleges have an open admission policy?
```

```python
# The key to understand Open admission Policy is the variable OPENADMP
# In class We decided that 1 means Open Policy 
Admis = MAC270['OPENADMP']
```

```python
Admis.head(8)
```

```python
# Since we have missing values we have to drop them
OpenAdmis = MAC270.dropna(subset=['OPENADMP'])
```

```python
# We decided that 2 means not open admissions
# We drop them since we don't need them for our calculation
openAdmis = openAdmis[openAdmis.OPENADMP !=2]
```

```python
# calculate the percentage of open admission out of all schools
print('The percentage of colleges that have open admission is:', round((len(openAdmis))/(len(MAC270))*100),
      '%')
```
